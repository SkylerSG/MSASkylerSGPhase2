```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
examscores = pd.read_csv("datasets\exam_scores.csv", delimiter=',')
pvexam = examscores.head(10)
print('\nFirst 10 lines, Exam Set:\n\n',pvexam)
```

```python
exmnum = examscores[["NrSiblings","MathScore","ReadingScore","WritingScore"]]
print("\nExam Score Mean(s):\n",exmnum.mean(),"\nExam Score Standard Deviation(s):\n",exmnum.std())
```

```python
examscores_boxplottable = examscores[["MathScore","ReadingScore","WritingScore"]]
plt.boxplot(examscores_boxplottable)  ## BOX PLOTS OF NUMERICAL VALUES OF EXAM SCORES (exmnum)
plt.ylabel("Score (out of 100)")
plt.show()
plt.hist(examscores["NrSiblings"])
plt.xlabel("Number of Siblings")
plt.ylabel("Frequency")
plt.show()
#print(examscores["NrSiblings"])
```
