```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
examscores = pd.read_csv("datasets\exam_scores.csv", delimiter=',')
mktsegmentation = pd.read_csv("datasets\market_segmentation.csv", delimiter=',')
pvexam = examscores.head(10)
pvmkt = mktsegmentation.head(10)
print('\nFirst 10 lines, Exam Set:\n\n',pvexam,'\n\nFirst 10 lines, Market Segmentation Set:\n\n',pvmkt)
```

```python
exmnum = examscores[["NrSiblings","MathScore","ReadingScore","WritingScore"]]
mktnum = mktsegmentation[["Age","Family_Size"]]

print("\nExam Score Mean(s):\n",exmnum.mean(),"\nMarket Segmentation Mean(s):\n",mktnum.mean(),"\nExam Score Standard Deviation(s):\n",exmnum.std(),"\nMarket Segmentation Standard Deviation(s):\n",mktnum.std())
```

```python
examscores_boxplottable = examscores[["MathScore","ReadingScore","WritingScore"]]
plt.boxplot(examscores_boxplottable)  ## BOX PLOTS OF NUMERICAL VALUES OF EXAM SCORES (exmnum)
plt.show()
plt.hist(examscores["NrSiblings"])
plt.show()
#print(examscores["NrSiblings"])
```

```python
market_boxplottable = mktsegmentation[["Age"]]
plt.boxplot(market_boxplottable)
plt.show()
plt.hist(mktnum["Family_Size"]) 
plt.show()
```
